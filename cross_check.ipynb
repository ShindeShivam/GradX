```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

np.random.seed(42)
    
# Generate non-linear data: y = 3x^2 + 2x + 1 + noise
X = np.linspace(-3, 3, 100)
X = X.reshape(-1,1)
y = 3 * X**2 + 2 * X + 1 + np.random.randn(100) * 2


test_X = np.array([-2, 0, 2])
test_X = test_X.reshape(-1,1)

poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X)
test_X = poly.transform(test_X)

model = LinearRegression()
model.fit(X_poly, y)
```

```python
model.predict(test_X)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

np.random.seed(42)

# 1. Generate data
X = np.linspace(-3, 3, 100).reshape(-1,1)
y = 3 * X**2 + 2 * X + 1 + np.random.randn(100) * 2

# 2. Define test points
test_X = np.array([[-2], [0], [2]])

# 3. Polynomial features
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X)
test_X_poly = poly.transform(test_X)

# 4. Train model
model = LinearRegression()
model.fit(X_poly, y)

# 5. Predict on test points
predictions = model.predict(test_X_poly)
print(predictions)
```
