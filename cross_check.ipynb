```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

np.random.seed(42)

# 1. Generate data
X = np.linspace(-3, 3, 100).reshape(-1,1)
y = 3 * X**2 + 2 * X + 1 + np.random.randn(100) * 2

# 2. Define test points
test_X = np.array([[-2], [0], [2]])

# 3. Polynomial features
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X)
test_X_poly = poly.transform(test_X)

# 4. Train model
model = LinearRegression()
model.fit(X_poly, y)

# 5. Predict on test points
predictions = model.predict(test_X_poly)
print(predictions)
```

## Training Neural Network on MNIST

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# Load full MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist["data"], mnist["target"].astype(int)

# Normalize pixel values
X = X / 255.0

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=5000, train_size=50000, random_state=42)

# One-hot encode labels
encoder = OneHotEncoder(sparse_output=False)
y_train_onehot = encoder.fit_transform(y_train.reshape(-1,1))
y_test_onehot = encoder.transform(y_test.reshape(-1,1))
```

### SGD

```python
from neural_network import NeuralNetwork 
```

```python
nn = NeuralNetwork(layer_sizes=[784, 64, 10], activation='relu', optimizer='sgd' )
```

```python
nn.fit(X_train, y_train_onehot, learning_rate=0.01, batch_size=32, epochs=20)
```

```python
y_pred_proba = nn.predict_proba(X_test) # Predict probabilities
y_pred = np.argmax(y_pred_proba, axis=1) # Pick class with highest probability
y_true = np.argmax(y_test_onehot, axis=1)

# Accuracy
accuracy = np.mean(y_pred == y_true)
print("Test Accuracy:", accuracy)
```
